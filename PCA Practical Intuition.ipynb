```python
from sklearn.decomposition import PCA
```

```python
from sklearn.datasets import load_iris
```

```python
import matplotlib.pyplot as plt
```

```python
dataset = load_iris()
```

```python
X = dataset["data"]
y = dataset["target"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
## PCA is a transformation technique
pca = PCA(n_components=3) ## x_train have 4 features, we are converting 4 features into 3 features.
```

```python
X_train = pca.fit_transform(X_train)
X_train
```

```python
pca.explained_variance_ratio_  ## How much data am i able to extract from the whole !!
```

```python
pca.components_  ## as we have transformed it from 4 to 3D... so these 3 are the coordinated of PC1, PC2 & PC3...
## tells how each original feature contributes to a PC
```
