##  Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.
## Answer

```python
df.head()
```

### There can be many ways :
#### 1. Remove the row or columns having missing values
- Advantage : Easy to implement.
- Disadvantage : This can cause intense data loss and get the performance of our model down.
#### 2. Replacing mode with the missing values
- Advantage : Simple and maintains sample size
- Disadvantage : Ignores feature interactions
#### K-Nearest Neighbors (KNN) Imputation:
- Predicts missing values based on neighboring data points.
- Advantage: Considers feature relationships.
- Disadvantage: Computationally expensive.

## 5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to these features to improve normality?
## Answer

```python
df.head()
```

```python
features = df.columns.tolist()
for feature in features:
    plt.figure(figsize=(7,2))
    sns.kdeplot(data=df, x=feature)
    plt.xlabel(feature)
    plt.show()
```

#### We can use many transformations to imprive the normality of the feature in a dataframe :

## Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler()
```

```python
scaled_data = minmax.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
data = instance.fit_transform(scaled_data)
```
