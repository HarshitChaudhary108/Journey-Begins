```python
df["Fare"] = np.random.randint(10, 300, size=1309)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
df["Parch"] = np.random.randint(0, 1, size=1309)
df["Embarked_C"] = oe.fit_transform(df[["Embarked_C"]])
df["Embarked_Q"] = oe.fit_transform(df[["Embarked_Q"]])
df["Embarked_S"] = oe.fit_transform(df[["Embarked_S"]])
```

```python
from scipy.stats import zscore
```

```python
X = df.drop("Survived", axis = 1)
y = df["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ab = AdaBoostClassifier(n_estimators=50, learning_rate=0.1, random_state=0)
ab.fit(X_train, y_train)

y_pred = ab.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb = GradientBoostingClassifier(n_estimators=50, learning_rate=0.1, random_state=0)
model = gb.fit(X_train, y_train)

pred1 = model.predict(X_test)
```

```python
accuracy_score(pred1, y_test)
```
