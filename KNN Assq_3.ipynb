## Q1. Write a Python code to implement the KNN classifier algorithm on load_iris dataset in sklearn.datasets.
## Answer 

```python
import seaborn as sns
df = sns.load_dataset("iris")
X = df.drop("species", axis=1)
y = df["species"]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

k_value = list(range(5,8))
param_s = {
    "n_neighbors" : k_value,
    "weights" : ["uniform", "distance"],
    "algorithm" : ["auto"],
    "p" : [1, 2]
}
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
hpt = GridSearchCV(estimator=KNeighborsClassifier(), param_grid=param_s, cv=3, scoring="accuracy")
hpt.fit(X_train, y_train)

hpt.best_params_
classifier = KNeighborsClassifier(n_neighbors=6, p=2, weights="uniform", algorithm="auto")
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy_score(y_pred, y_test)
```

## Q2. Write a Python code to implement the KNN regressor algorithm on load_boston dataset in sklearn.datasets.
## Answer 

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
raw_df = fetch_california_housing()
df = pd.DataFrame(raw_df.data, columns=raw_df.feature_names)
df1 = pd.DataFrame(raw_df.target, columns=raw_df.target_names)
X = df
y = df1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
from sklearn.neighbors import KNeighborsRegressor
param_s = {
    "n_neighbors" : list(range(5,30)),
    "weights" : ["uniform", "distance"],
    "algorithm" : ["auto"],
    "p" : [1,2]
}

hpt = GridSearchCV(estimator=KNeighborsRegressor(), cv=3, param_grid=param_s)
hpt.fit(X_train, y_train)
hpt.best_params_
regressor = KNeighborsRegressor(n_neighbors=10, p=1, weights="distance", algorithm="auto")
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

## Q3. Write a Python code snippet to find the optimal value of K for the KNN classifier algorithm using cross-validation on load_iris dataset in sklearn.datasets.
## Answer 

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
df = load_iris()
X,y = df.data, df.target
```

```python
Kvalue = list(range(1,30))
cv_score = []
```

```python
for k in Kvalue:
    classifier = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(classifier, X, y, scoring="accuracy", cv=5)
    cv_score.append(scores.mean())
```

```python
## to find optimal value of k for the model
import numpy as np
optimal_k = Kvalue[np.argmax(cv_score)] ## [np.argmax(cv_score) = the index of highest accuracy in cv_score.
print(optimal_k)
print(max(cv_score))
```

## Q6. Implement a function to standardise the features before applying KNN classifier.
## Answer 

```python
def KNN_Standardiser(X_train, X_test, y_train, y_test, n_neighbors=5, weights="uniform", p=2):
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    model = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights, p=p)
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_pred, y_test)
    return accuracy
```

## Q7. Write a Python function to calculate the euclidean distance between two points.
## Answer

```python
import math
def euclidean_distance(point1, point2):
    if len(point1) != len(point2):
        print("The dimension of the points are not same.")
    else:
        return math.sqrt(sum((a - b) ** 2 for a, b in zip(point1, point2)))
```

```python
euclidean_distance([3,4], [7,1])
```

## Write a Python function to calculate the manhattan distance between two points.
## Answer 

```python
import math
def manhattan_distance(point1, point2):
    if len(point1) != len(point2):
        print("The dimension of the points are not same.")
    else:
        return sum(abs(a - b) for a, b in zip(point1, point2))
```

```python
manhattan_distance([3,1], [5,3])
```
